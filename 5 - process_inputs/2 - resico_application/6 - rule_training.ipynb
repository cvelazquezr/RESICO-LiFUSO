### Analysing the unclassified dataset *with* import statements and matching them with the predicted library to already establish a classification

```python
for index_row, row in unclassified_imports.iterrows():
    imports = list(set(row['Imports'].split(",")))
    predicted_libraries = row['Predicted_Library'].split(",")
    
    in_imports = False
    
    for predicted_import in predicted_libraries:
        for import_stmt in imports:
            if import_stmt.startswith(predicted_import):
                in_imports = True
    
    if in_imports:
        unclassified_imports.at[index_row, 'Library_Used'] = 'Y'
    else:
        unclassified_imports.at[index_row, 'Library_Used'] = 'N'
```

### Rules (based on the imbalanced dataset)

The rules perform better in the imbalanced scenario, probably because they capture better the characteristics of an instance being predicted as Y or N, even though the imbalance of the data. This behaviour has been reported by at least two papers:

- The Impact of Class Rebalancing Techniques on the Performance and Interpretation of Defect Prediction Models by Chakkrit Tantithamthavorn; Ahmed E. Hassan; Kenichi Matsumoto
- Dealing with imbalanced data for interpretable defect prediction by Yuxiang Gao, Yi Zhu, Yu Zhao

```python
def j48(clas, meth, amb):
    if amb == 0:
        if meth == 0:
            return 'N'
        return 'Y'
    else:
        if clas <= 4:
            if meth == 0:
                return 'N'
            else:
                if amb <= 1:
                    if clas <= 2:
                        return 'N'
                    return 'Y'
                return 'N'
        else:
            if amb <= 2:
                return 'Y'
            else:
                if clas <= 8:
                    if amb <= 3:
                        if clas <= 6:
                            return 'N'
                        return 'Y'
                    return 'N'
                else:
                    if amb <= 7:
                        if amb <= 5:
                            return 'Y'
                        else:
                            if clas <= 13:
                                return 'N'
                            return 'Y'
                    else:
                        if clas <= 18:
                            return 'N'
                        else:
                            if meth <= 19:
                                return 'Y'
                            return 'N'
```

### Inspecting the predictions made by the rule model

```python
hits = 0
fails = 0

for _, row in data_processed.iterrows():
    prediction = j48(row.Classes, row.Methods, row.Ambiguity)
    
    if prediction == row.Library_Used:
        hits += 1
    else:
        fails += 1

print("Hits: {} Failures: {}".format(hits, fails))
print("Hits accuracy: {}".format(round(hits / (hits + fails), 2)))
```

```python
selected_index_failures = []

for index_row, row in data_processed.iterrows():
    prediction = j48(row.Classes, row.Methods, row.Ambiguity)
    real_prediction = row.Library_Used 
    
    if prediction != real_prediction and real_prediction == 'Y':
        selected_index_failures.append(index_row)
```
